```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

fig,axes = plt.subplots(2,3,tight_layout=True)
# 折れ線グラフ
y0 = [1,2,-5,2]
axes[0,0].set_title("折れ線グラフ")
axes[0,0].plot(y0)
# sinカーブ
x1 = np.linspace(0,2*np.pi,50)
y1 = np.sin(x1)
axes[0,1].set_title("sinカーブ")
axes[0,1].plot(x1,y1)
# ヒストグラム
x2 = np.random.normal(0.5,0.3,200)
axes[1,0].set_title("ヒストグラム")
axes[1,0].hist(x2,bins=30) # bins:棒の数
# 散布図
#np.random.seed(101)
x3 = np.random.normal(0.5,0.2,60)
y3 = np.random.normal(0.5,0.2,len(x3))
axes[1,1].set_title("散布図")
axes[1,1].scatter(x3,y3)

plt.show()
```
